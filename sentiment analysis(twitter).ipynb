## Making all words to lower case

```python
df['new']=df['new'].apply(lambda x:" ".join([x.lower() for x in x.split()]))
df.head(100)
```

## making all text into vectorized form


```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfi=TfidfVectorizer(max_features=1000,lowercase='True',analyzer='word',stop_words='english',ngram_range=(1,3))
v=tfi.fit_transform(df['new'])
v.toarray()
```

```python
x=v.toarray()
y=df['label']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=0)
```

## Naive_bayes algo applied for sentiment analysis

```python
from sklearn.naive_bayes import MultinomialNB
mnb=MultinomialNB()
mnb.fit(x_train,y_train)
```

```python
y_pred=mnb.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
from sklearn.metrics import accuracy_score
print('accuracy_score',accuracy_score(y_test,y_pred))
```

```python
prob=mnb.predict_proba(x_test)
prob
```

```python
from sklearn import metrics
prob_positive=prob[:,1]
fpr,tpr,threshold=metrics.roc_curve(y_test,prob_positive)
roc_auc=metrics.auc(fpr,tpr)
print("Area under the curve:",roc_auc)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
plt.rcParams['figure.figsize']=6,5
plt.rcParams['axes.facecolor'] = 'cyan'
plt.title('Receiver opertating Characteristics')
plt.plot(fpr,tpr,'orange',label='AUC = % .2f' % roc_auc)
plt.legend(loc='lower right')
plt.plot([0,1],[0,1],color='darkblue',linestyle='--')
plt.ylabel('True positive rate')
plt.xlabel('False positive rate')
plt.xticks()
plt.show()
```

## Textblob sentiment analysis

```python
demo=df['tweet'][:5].apply(lambda x: TextBlob(x).sentiment )
demo
#here the tuple contains polarity and subjectivity(polarity 1 means +ve -1 means -ve)
```

```python
df['sentiment'] = df['tweet'].apply(lambda x: TextBlob(x).sentiment[0] )
```

```python
df.head(2)
```

```python
df['pred']=df.sentiment.apply(lambda x: 0 if x>=0 else 1)
```

```python
df.head(10)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
cm=confusion_matrix(df.label,df.pred)
print(cm)
print('accuracy_score=',accuracy_score(df.label,df.pred))
```
